# Global Economic Crisis Analysis
## How Banking Collapses, Currency Crashes, and Sovereign Debt Defaults Reshape Economies Across Income Groups (2000–2024)

**Dataset:** 26,000 country-year observations spanning 80 countries across 4 income groups  
**Period:** 2000–2024  


# Problem Statement
**Persistent Economic Instability**: Global economies frequently face systemic shocks, including banking collapses, currency crashes, and sovereign debt defaults.

**Data Fragmentation**: There is a lack of integrated analysis that links financial crisis types with their direct impact on socio-economic indicators like poverty and human development.

**Income Group Disparity**: It is unclear how crisis outcomes and recovery durations differ significantly between High, Middle, and Low-income nations.

**Compound Risks**: The triggers and consequences of "twin" or "triple" crises (simultaneous banking, debt, and currency failures) are not sufficiently quantified.

**Predictive Gaps**: Traditional metrics often fail to identify early warning signs (like NPL ratios or stock market volatility) across diverse regional contexts.

# Problem Objectives
**Quantify Macroeconomic Impact**: Measure the specific loss in GDP growth and spikes in inflation caused by different types of financial crises.

**Analyze Income Inequality**: Compare how crises affect poverty rates and Gini coefficients across different income classifications.

**Identify Leading Indicators**: Determine which variables (e.g., Debt-to-GDP, Foreign Reserves, or NPL ratios) serve as the most reliable predictors of sovereign default.

**Evaluate Recovery Timelines**: Calculate the average number of years required for different economies to return to pre-crisis growth levels.

**Assess Systemic Linkages**: Investigate the correlation between capital flight, currency depreciation, and the subsequent need for IMF intervention programs.

**Develop Policy Recommendations**: Provide data-driven insights for building resilient financial buffers and social protection floors based on historical crisis patterns.

## Data Cleaning & Preprocessing

We handle missing values as follows:
- **Numeric columns**: fill with median (robust to outliers)
- **Crisis-only columns** (`crisis_name`, `crisis_severity`, `recovery_years`): fill with `'No Crisis'` / `'None'`  since nulls just mean no crisis occurred
- **Derived features**: create `crisis_status` and `default_status` helper columns

```python
df_clean = df.copy()

# fill numeric columns with median
num_cols = df_clean.select_dtypes(include='number').columns
df_clean[num_cols] = df_clean[num_cols].fillna(df_clean[num_cols].median())

# fill crisis-only string columns
df_clean['crisis_name'] = df_clean['crisis_name'].fillna('No Crisis')
df_clean['crisis_severity'] = df_clean['crisis_severity'].fillna('None')

# ordered income group category
income_order = ['Low Income', 'Lower Middle Income', 'Upper Middle Income', 'High Income']
df_clean['income_group'] = pd.Categorical(df_clean['income_group'], categories=income_order, ordered=True)

print('Missing values after cleaning:', df_clean.isnull().sum().sum())
df_clean.describe()
```

- All numeric nulls filled with column medians — no rows dropped.
- `crisis_name` and `crisis_severity` filled with `'No Crisis'` / `'None'` for non-crisis rows.
- Dataset is now clean and ready for visualization.

## Data Visualization


### Q1: How does GDP growth differ across income groups?

```python
# average GDP growth by income group
gdp_by_income = df_clean.groupby('income_group')['gdp_growth_pct'].mean().reindex(income_order)

plt.figure(figsize=(10, 5))
sns.barplot(x=gdp_by_income.index, y=gdp_by_income.values, color='#4287f5')
plt.title('Average GDP Growth by Income Group', fontsize=20)
plt.xlabel('Income Group', fontsize=15)
plt.ylabel('Avg GDP Growth (%)', fontsize=15)
plt.xticks(rotation=15)
plt.show()
```

- Lower and Upper Middle Income countries show the highest average GDP growth.
- High Income countries grow more slowly but more stably.

### Q2: Which regions experienced the most banking crises?

```python
# count of banking crisis events by region
crisis_by_region = df_clean[df_clean['banking_crisis'] == 1].groupby('region').size().sort_values(ascending=False)

plt.figure(figsize=(10, 5))
ax = sns.barplot(x=crisis_by_region.values, y=crisis_by_region.index, color='#4287f5')
ax.bar_label(ax.containers[0], label_type='edge')
plt.title('Banking Crisis Count by Region', fontsize=20)
plt.xlabel('Number of Crisis Country-Years', fontsize=15)
plt.ylabel('Region', fontsize=15)
plt.show()
```

- Europe and the Americas had the highest banking crisis frequency, largely driven by the 2008 Global Financial Crisis.
- Sub-Saharan Africa shows fewer formal banking crises despite economic challenges.

### Q9: How do key economic indicators correlate with each other?


```python
# correlation heatmap of key variables
key_cols = ['gdp_growth_pct', 'inflation_rate_pct', 'unemployment_rate_pct',
            'govt_debt_pct_gdp', 'poverty_rate_pct', 'gini_coefficient',
            'hdi_score', 'banking_crisis', 'sovereign_default', 'currency_crash']

plt.figure(figsize=(10, 8))
corr = df_clean[key_cols].corr()
mask = np.triu(corr)
sns.heatmap(corr, annot=True, mask=mask, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Between Key Economic Indicators', fontsize=16)
plt.show()
```

### Q20: How does stock market volatility differ across crisis types?


```python
# Stock market returns across crisis types
crisis_type_map = {
    'No Crisis': (df_clean['banking_crisis'] == 0) & (df_clean['currency_crash'] == 0) & (df_clean['sovereign_default'] == 0),
    'Banking Crisis': df_clean['banking_crisis'] == 1,
    'Currency Crash': df_clean['currency_crash'] == 1,
    'Sovereign Default': df_clean['sovereign_default'] == 1
}

stock_data = []
for label, mask in crisis_type_map.items():
    sub = df_clean.loc[mask, 'stock_market_return_pct'].dropna()
    stock_data.append(pd.DataFrame({'Crisis Type': label, 'Stock Return (%)': sub}))
stock_df = pd.concat(stock_data)

plt.figure(figsize=(10, 5))
sns.boxplot(
    x='Crisis Type', y='Stock Return (%)', data=stock_df,
    palette=['#4287f5', '#f54242', '#f5a742', '#a742f5']
)
plt.title('Stock Market Returns by Crisis Type', fontsize=20)
plt.xlabel('Crisis Type', fontsize=15)
plt.ylabel('Stock Market Return (%)', fontsize=15)
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')
plt.show()
```

- Stock market returns turn sharply negative during all three crisis types.
- Currency crashes produce the widest spread in returns, reflecting market uncertainty.
- Sovereign defaults show the most consistently negative returns, as investor confidence collapses.

In this notebook, we explored the Global Economic Crisis dataset (2000–2024) and drew the following key insights:

- **Income group matters**: Lower and Upper Middle Income countries grow faster but face higher crisis volatility. High Income countries grow slowly but stably.
- **Crises hurt everywhere**: Banking crises suppress GDP growth and raise unemployment across all income groups, with the most severe impact on already-vulnerable economies.
- **Inflation spikes in crises**: Two major global spikes (2008 GFC, 2021–23 post-COVID) are clearly visible, with Low Income countries chronically suffering higher baseline inflation.
- **Debt and default are linked**: Countries that default carry significantly higher debt loads, though governance and reserves also play key roles.
- **Poverty and HDI gap is stark**: Low Income countries have poverty rates exceeding 60% and HDI scores below 0.5, compared to near-zero poverty and ~0.9 HDI in High Income countries.
- **Crises are contagious**: Banking crises, currency crashes, and sovereign defaults tend to occur together (twin/triple crises), as confirmed by the correlation heatmap and triple crisis frequency analysis.
- **IMF programs help recovery**: Countries under IMF programs show slightly faster and more consistent recovery timelines. Low Income nations depend on them most heavily.
- **Inequality is rising in rich countries**: Post-2008 inequality increased in High Income countries, while Low Income countries maintain persistently high Gini coefficients.
- **Capital flight fuels currency crashes**: Higher capital outflows are tightly correlated with severe currency depreciation, highlighting the self-reinforcing nature of financial panic.
- **NPL ratios signal banking stress**: Non-performing loan ratios spike sharply before and during banking crises, confirming their role as a critical early-warning indicator.
- **Stock markets crash across all crisis types**: Returns turn negative under banking crises, currency crashes, and sovereign defaults alike, with currency and default events producing the most severe and volatile market reactions.
